# Create a list of samples in which you vary solute concentrations from 0 to 200 mM.

```python
# Sample code

## INITIALIZATION
all_samples = []

## LOAD INFORMATION ABOUT THE STOCK SOLUTIONS
stock_dict = sh.read_stocks(filepath='./preparation/stocks.csv')

# For this exercise, we will use three amino acids as solutes.
solutes = ['Arginine', 'Lysine', 'Proline',]
# Concentrations of the samples that will be created:
concentrations = [0, 10, 25, 50, 100, 200] # in mM

for solute in solutes:
    # LOAD INFORMATION ABOUT THE STOCK SOLUTIONS EVERY LOOP
    ## ... THIS IS TO REFLECT THE CHANGES IN THE VOLUME OF STOCK STOLUTIONS DURING EXPERIMENTS
    stock_dict = sh.read_stocks(filepath='./preparation/stocks.csv')
    
    if solute in ['Phenylalanine', 'Glutamic_acid']:
        concentrations = [10, 20, 30, 40,] # Solute concentrations for species with low solubility in water
    else:
        concentrations = [0, 10, 25, 50, 100, 200] # Solute concentrations for species with high solubility in water

    # CREATE A SAMPLE LIST
    samples = sh.create_samples(composition = {'Bik1': 1.8, 'Bik1_buffer': 19, solute: 0, 'water': 0}, # water is always zero
                             vary=solute, # 
                             concentrations= concentrations, 
                             solvent='water', # 
                             order_mixing=['Bik1', [solute, 'Bik1_buffer', 'water']],
                             special_species=['Bik1'], # Species that require special mixing protocol (e.g. viscous samples)
                             skip_species = ['Bik1', solute], # Species that will be skipped during mixing 
                             stock_dict=stock_dict, volume=40, # Provide with stock_dict and sample volume to create a recipe
                             name='bik1_' + solute,
                             repeat=1, 
                             save=True, 
                            )
    all_samples.append(samples)

# STEP 3: Create a labware configuration file
# sh.create_default_labware_configuration(all_samples)

# STEP 4: USER MUST EDIT THE FILE at /preparation/lab_configuration.csv

# STEP 5: LOAD the labware configuration file
dfs = sh.read_labware_configuration()

# STEP 6: Assign a unique well position to each sample
sh.assignWellPositions(all_samples, )

# STEP 7: (Optional) Create a sample sheet (Sample name, well number, composition, etc.)
sh.create_sample_info(all_samples)

# # # STEP 8: Upload the file to Otto (server)
sp.upload_exerimental_protocol()

# # # STEP 9: Check how the robot would run.
!opentrons_simulate startProtocol.py -e -o nothing > ./archive/expected_output.txt

speak("Done")
```
